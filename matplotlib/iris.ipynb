```python
from sklearn import datasets
import numpy as np
#Load the numpy library for fast array computations
import pandas as pd
#Load the pandas data-analysis library
import matplotlib.pyplot as plt
#Load the pyplot visualization library
```

```python
iris = datasets.load_iris()
#iris data
iris.data
```

```python
#target names 
iris.target_names
```

```python
iris_df = pd.DataFrame(iris.data, columns= iris.feature_names)
```

```python
# sepal length histogram
iris_df['sepal length (cm)'].hist(bins=30)
```

```python
# histogram by target variable(coloured)
for class_name in np.unique(iris.target):
    plt.figure(1)
    iris_df['sepal length (cm)'].iloc[np.where(iris.target == class_name)[0]].hist(bins=30)
```

```python
data = iris.data
target = iris.target
# Resize the figure for better viewing
plt.figure(figsize=(12,5))
# First subplot
plt.subplot(121)
# Visualize the first two columns of data:
#c = target --> takes an array of colorrs  i.e 0,1,2 in target 
plt.scatter(data[:,0], data[:,1], c=target)
# Second subplot
plt.subplot(122)
# Visualize the last two columns of data:
plt.scatter(data[:,2], data[:,3], c=target)
```
