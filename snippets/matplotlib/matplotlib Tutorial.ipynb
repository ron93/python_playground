# Matplotlib Tutorial

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

## Plot

```python
x_1 = np.linspace(0, 5,10)
y_1 = x_1**2
plt.plot(x_1,y_1)
plt.title('Days squared chart')
plt.ylabel('days Squared')
plt.xlabel('Days')
```


## Multiple plots

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1,'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_1,'b')
```

## Figure object

```python
fig_1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig_1.add_axes([0.1, 0.1,0.9,0.9])
axes_1.set_xlabel('Days')
axes_1.set_xlabel('Days squared')
axes_1.set_title('Days squared plot')
axes_1.plot(x_1,y_1, label='x/x2')
axes_1.plot(y_1,x_1, label='x2/x')
# upper-right: 1 , upper-left: 2, lower-left: 3, lower-right:4,
#or tuple of x & y from lower-left(0,0), upper-left(0,1) etc.
#0 : places legend in best possible possition
axes_1.legend(loc=0)

axes_2 = fig_1.add_axes([0.45,0.45,0.4,0.3])
axes_2.set_xlabel('Days')
axes_2.set_xlabel('Days squared')
axes_2.set_title('Days squared plot')
axes_2.plot(x_1,y_1, 'r')

axes_2.text(0,40,'Ploting')
```

## subplots

```python
fig_2, axes_2 = plt.subplots(figsize=(8,4), nrows=1,ncols=3)
plt.tight_layout()

axes_2[0].set_title('Plot 1')
axes_2[0].set_ylabel('X')
axes_2[0].plot('X squared')
axes_2[0].plot(x_1,y_1)

axes_2[1].set_title('Plot 2')
axes_2[1].set_ylabel('X')
axes_2[1].plot('X squared')
axes_2[1].plot(x_1,y_1)

axes_2[2].set_title('Plot 3')
axes_2[2].set_ylabel('X')
axes_2[2].plot('X squared')
axes_2[2].plot(x_1,y_1)
```

# plotting using a function

```python
import numpy as np
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

# Calling my_plotter function

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})

#with two sub-plots
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```
